# hierarchical clustering

there are two types of hc clustering

1)Agglomerative (bottom up apporach)

2)Divisive



# Agglomerative 
# steps 

1)make each data point a single point cluster that forms n clusters

2)take the two closet data points and make them one cluster,that forms n-1 clusters

3)take the two closet clustes and make them one cluster,that forms n-2 clusters

4)repeat step 3 until there is only one cluster

note:we use here closet distance between two cluster

# Distance between two clusters

1)closet points

2)furthest points 

3)average distance 

4)Distance between centroids

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("./Mall_Customers.csv")
```

```python
features=dataset.iloc[:,[3,4]].values
```

```python
# dendogram
import scipy.cluster.hierarchy as sch
dendogram=sch.dendrogram(sch.linkage(features,method='ward'))
plt.xlabel("customer")
plt.ylabel("expense")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
```

```python
y_hc=model.fit_predict(features)
```

```python
plt.scatter(features[y_hc == 0, 0], features[y_hc == 0, 1], s = 100, c = 'red', label = 'Cluster 1')
plt.scatter(features[y_hc == 1, 0], features[y_hc == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(features[y_hc == 2, 0], features[y_hc == 2, 1], s = 100, c = 'green', label = 'Cluster 3')
plt.scatter(features[y_hc == 3, 0], features[y_hc == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(features[y_hc == 4, 0], features[y_hc == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
